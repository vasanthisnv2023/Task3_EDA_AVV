```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
df.head()
```

```python
df.info()
df.describe()
```

The Iris dataset contains numerical features related to flower measurements and a categorical target variable representing species.

```python
df.hist(figsize=(10,8))
plt.show()
```

Histograms show that petal features have more variation and better separation among species.

```python
sns.countplot(x='species', data=df)
plt.show()
```

Each species has an equal number of samples, indicating a balanced dataset.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop('species', axis=1))
plt.show()
```

Box plots reveal potential outliers in sepal length and width features.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.show()
```

Petal length and petal width show strong positive correlation, making them important predictive features.

Based on EDA, petal length and petal width are the most informative features for species classification.

Petal-related features show strong correlation

Dataset is balanced across species

Some outliers exist but do not significantly affect data

Petal length and width are important features for prediction
